```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib
```

```python
data = pd.read_csv('campus_crime.csv')
```

```python
data[:1]
```

```python
data.sort_values(by='total_enr', ascending=False)[:10]
```

```python
data.info()
```

```python
##counting all types of crimes for each college
data["total count"] = data["murder"] + data["manslaughter"] + data["forcible"] + data["nonforce"] + data["robbery"] + data["agg_aslt"] + data["burglary"] + data ["vehicle"] + data["arson"]
data.sort_values(by='total count', ascending=False)[:10]
```

```python
##number of crimes per 1,000 students at each college
data['crime_pct'] = (data['total count']/data ['total_enr'])*1000
data.sort_values(by='crime_pct', ascending=False)
```
